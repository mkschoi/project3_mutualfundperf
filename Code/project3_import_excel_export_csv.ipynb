### 1. Import excel files containing raw mutual fund data

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('../Data/US_mutual_fund_data.xlsx')
df2 = pd.read_excel('../Data/US_mutual_fund_data2.xlsx')
```

```python
df1.head()
```

### 2. Align the columns and drop empty rows

```python
df1.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df1.drop(df1[df1.isnull().all(1)].index, inplace=True)
```

```python
df1
```

```python
df2.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df2.drop(df2[df2.isnull().all(1)].index, inplace=True)
```

```python
df2
```

```python
df1[df1['Unnamed: 1'].notnull()] = df1[df1['Unnamed: 1'].notnull()].shift(1, axis=1)
```

```python
df1.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 
          'Unnamed: 12', 'Unnamed: 14', 'Unnamed: 16', 'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 22'], axis=1, inplace=True)
```

```python
df1.rename(columns={'Unnamed: 3':'fund_name'}, inplace=True)
```

```python
df1
```

```python
df2[df2['Unnamed: 1'].notnull()] = df2[df2['Unnamed: 1'].notnull()].shift(1, axis=1)
```

```python
df2.drop(['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 
          'Unnamed: 12', 'Unnamed: 14', 'Unnamed: 16', 'Unnamed: 18', 'Unnamed: 20', 'Unnamed: 22'], axis=1, inplace=True)
```

```python
df2.rename(columns={'Unnamed: 3':'fund_name'}, inplace=True)
```

```python
df1.reset_index(inplace=True, drop=True)
```

```python
df2.reset_index(inplace=True, drop=True)
```

```python
df1
```

```python
df2
```

### 3. Add Mutual Fund Tickers

```python
df3 = pd.read_excel('../Data/mutual_fund_tickers.xlsx')
```

```python
df4 = pd.concat([df3,df1],axis=1)
```

```python
df4.drop('fund_name',axis=1,inplace=True)
```

```python
df4
```

```python
df5 = pd.concat([df3,df2],axis=1)
```

```python
df5.drop('fund_name',axis=1,inplace=True)
```

```python
df5
```

### 4. Export the dataframes as CSV files

```python
df4.to_csv('../Data/US_mutual_fund_data_left.csv', index=False)
df5.to_csv('../Data/US_mutual_fund_data_right.csv', index=False)
```
