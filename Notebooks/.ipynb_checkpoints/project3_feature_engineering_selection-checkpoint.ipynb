## Feature Engineering

### Convert categorical features to dummy variables

```python
no_load = pd.get_dummies(df_mf['No-load Funds'], drop_first=True)
closed = pd.get_dummies(df_mf['Closed to New Investors'], drop_first=True)
```

```python
df_mf.drop(['No-load Funds','Closed to New Investors'], axis=1, inplace=True)
```

```python
df_mf = pd.concat([df_mf, no_load, closed], axis=1)
```

### Encode ordinal categorial feature (Morningstar Risk)

```python
scale_mapper = {'Low':1, 'Below Average':2, 'Average':3, 'Above Average':4, 'High':5}
```

```python
df_mf2 = df_mf.copy()
```

```python
df_mf2['Morningstar Risk'] = df_mf2['Morningstar Risk'].replace(scale_mapper)
```

### Binning continuous features and other feature engineering techniques

## Feature selection

```python
X = df_mf2.drop(['Ticker','Outperform / Underperform'], axis=1)
y = df_mf2['Outperform / Underperform']
```

### Removing features with low variance

```python
sel_variance_threshold = VarianceThreshold()
```

```python
X_remove_variance = sel_variance_threshold.fit_transform(X)
print(X_remove_variance.shape)
```

### Univariate features selection

#### (1) Chi-square test

```python
sel_chi2 = SelectKBest(chi2, k=10)
```

```python
scaler_minmax = MinMaxScaler()
X_scaled = scaler_minmax.fit_transform(X)

X_chi2 = sel_chi2.fit_transform(X_scaled,y)
df_chi2 = pd.DataFrame(sel_chi2.get_support(), index=X.columns, columns=['Chi2'])
```

```python
df_chi2
```

#### (2) f test

```python
sel_f = SelectKBest(f_classif, k=10)
```

```python
X_f = sel_f.fit_transform(X,y)
df_f = pd.DataFrame(sel_f.get_support(), index=X.columns, columns=['f'])
```

```python
df_f
```

#### (3) mutual_info_classif test

```python
sel_mutual = SelectKBest(mutual_info_classif, k=10)
```

```python
X_mutual = sel_mutual.fit_transform(X,y)
df_mutual = pd.DataFrame(sel_mutual.get_support(), index=X.columns, columns=['f'])
```

```python
df_mutual
```
